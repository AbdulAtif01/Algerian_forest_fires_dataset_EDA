# Creating Dependent and Independent features

```python
df.columns
```

```python
x=pd.DataFrame(df, columns=['day', 'month', 'RH', 'WS', 'Rain', 'FFMC', 'DMC', 'DC',
       'ISI', 'BUI', 'FWI', 'region'])  

y=pd.DataFrame(df,columns=['Temperature'])
```

```python
x.head()
```

```python
y.head()
```

```python
#Train Test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=15)
```

```python
print(f'x_train.shape {x_train.shape}')
print(f'y_train.shape {y_train.shape}')
print(f'x_test.shape {x_test.shape}')
print(f'y_test.shape {y_test.shape}')
```

# Standardizing or Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()  ## Initialising
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
regression
regression.fit(x_train,y_train)
```

```python
## print the coefficients
print(regression.coef_)
```

```python
## print the intercept
print(regression.intercept_)
```

```python

## prediction for the test data
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
## Assumptions of Linear Regression


plt.figure(figsize=(8,8))
sns.regplot(x=y_test,y=reg_pred,data=df)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
## Checking the outliers of the target 'Temperature' feature
sns.boxplot(df['Temperature'])
```

```python
## Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
## R square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
## Adjusted R Square
#display adjusted R square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```

```python
#Ridege Regression Algorithm
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge
ridge.fit(x_train, y_train)
```

```python
## print the coefficients 
print(ridge.coef_)
```

```python

## print the intercept
print(ridge.intercept_)
```

```python
## prediction for the test data
ridge_pred = ridge.predict(x_test)
ridge_pred
```

```python
## residuals 
residuals = y_test-ridge_pred
residuals.head()
```

```python

#Lasso Regression Model Training
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso
lasso.fit(x_train,y_train)
```

```python
## print the coefficients
print(lasso.coef_)
```

```python
## print the intercept
print(lasso.intercept_)
```

```python
## prediction for the test data
lasso_pred = lasso.predict(x_test)
lasso_pred
```

```python
print(mean_squared_error(y_test, lasso_pred))
print(mean_absolute_error(y_test, lasso_pred))
print(np.sqrt(mean_squared_error(y_test, lasso_pred)))
```

```python
# R square
from sklearn.metrics import r2_score
score = r2_score(y_test, lasso_pred)
print(score)
```
